```python
# Task 2: Mark the success/failed launches for each site on the map
# Load SpaceX launch data (Ensure your dataset contains these fields)
spacex_df = pd.DataFrame({
    'LaunchSite': ['CCAFS LC-40', 'CCAFS LC-40', 'VAFB SLC-4E', 'VAFB SLC-4E', 'KSC LC-39A', 'KSC LC-39A'],
    'Lat': [28.562302, 28.563197, 34.632834, 34.632834, 28.573255, 28.573255],
    'Long': [-80.577356, -80.576820, -120.610745, -120.610745, -80.646895, -80.646895],
    'Class': [1, 0, 1, 0, 1, 0]  # 1 = Success, 0 = Failure
})

# Create a map centered around the average launch location
launch_map = folium.Map(location=[28.5, -80.6], zoom_start=5)

# Loop through each launch and add a marker
for index, row in spacex_df.iterrows():
    color = "green" if row["Class"] == 1 else "red"  # Green for success, red for failure
    
    folium.Marker(
        location=[row["Lat"], row["Long"]],
        popup=f"{row['LaunchSite']} - {'Success' if row['Class'] == 1 else 'Failure'}",
        icon=folium.Icon(color=color, icon="info-sign")
    ).add_to(launch_map)

# Display the map
launch_map
```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
import folium
from math import radians, cos, sin, asin, sqrt

# Haversine Formula to calculate distances between two lat/lon points
def haversine(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    km = 6371 * c  # Earth's radius in km
    return km

# Example Launch Site (CCAFS LC-40)
launch_site = (28.562302, -80.577356)
# Example Proximity Points
proximities = {
    "Railway": (28.5721, -80.5853),
    "Coastline": (28.5634, -80.5680),
    "City Center": (28.4014, -80.6043)
}

# Create a map centered on the launch site
site_map = folium.Map(location=launch_site, zoom_start=10)

# Add launch site marker
folium.Marker(launch_site, popup="CCAFS LC-40 (Launch Site)", icon=folium.Icon(color="blue", icon="rocket")).add_to(site_map)

# Loop through proximity points, calculate distances, and add to map
for name, coords in proximities.items():
    distance = haversine(launch_site[0], launch_site[1], coords[0], coords[1])
    folium.Marker(coords, popup=f"{name}: {distance:.2f} km", icon=folium.Icon(color="gray")).add_to(site_map)
    
    # Draw a line between the launch site and the proximity point
    folium.PolyLine([launch_site, coords], color="blue", weight=2.5, opacity=0.7).add_to(site_map)

# Display the map
site_map

```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
